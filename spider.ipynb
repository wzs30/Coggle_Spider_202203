## 任务3：bs4基础使用

学习资料：https://beautiful-soup-4.readthedocs.io/en/latest/
- 步骤1：使用requests和bs4爬取sklearn api页面
- 步骤2：在api页面中有多少个模块？有多少个API？如sklearn.base.DensityMixin，其中base为模块，DensityMixin为API
- 步骤3：将模块名作为key，api作为value存储为字典。

```python
# 导入相关库
import requests
from bs4 import BeautifulSoup
```

### 步骤1：使用requests和bs4爬取sklearn api页面

```python
url = "https://scikit-learn.org/stable/modules/classes.html"

resp = requests.get(url)
soup = BeautifulSoup(resp.text, 'html.parser')
soup
```

### 步骤2：在api页面中有多少个模块？有多少个API？如sklearn.base.DensityMixin，其中base为模块，DensityMixin为API

```python
# 找模块数：找section[@id="api-reference"]下所有id以“module”开头的section
import re

# 在子模块中找含"sklearn"的节点
modules = soup.find(id='api-reference').find_all(id=re.compile("sklearn"), recursive=False)
apis = soup.find_all('tr')

print(f"模块数：{len(modules)}\nAPI数：{len(apis)}")
```

### 步骤3：将模块名作为key，api作为value存储为字典。

```python
from collections import defaultdict

moduleDict = defaultdict(set)
for api in apis:
    st = api.find('span').string
    if '.' in st:
        ind = st.index('.')
        key, val = st[:ind], st[ind+1:]
        moduleDict[key].add(val)
    else:
        moduleDict['sklearn'].add(st)

moduleDict
```

## 任务4：bs4高阶使用

- 步骤1：爬取sklearn 机器学习名词页面, 爬取所有的名词，如1d、2d和api；
- 步骤2：有一些名词在介绍时，会有额外的链接，请将每个名词介绍对应的链接也找出。
- 步骤3：原始的名词安装本身有类别，如下所示，你能将爬取的结果进行分类吗？

```python
# 导入相关库
import requests
from bs4 import BeautifulSoup
```

### 步骤1：爬取sklearn 机器学习名词页面, 爬取所有的名词，如1d、2d和api；

```python
url = "https://scikit-learn.org/stable/glossary.html"

resp = requests.get(url)
soup = BeautifulSoup(resp.text, "html.parser")
```

```python
dts = soup.find_all('dt')
dtList = []
for dt in dts:
    for st in dt.strings:
        dtList.append(st)
        break

dtList
```

### 步骤2：有一些名词在介绍时，会有额外的链接，请将每个名词介绍对应的链接也找出。

```python
# 找出所有的站外链接
hrefList = []

aList = soup.find(id="glossary-of-common-terms-and-api-elements").find_all('a', class_ = re.compile("external"))
for a in aList:
    hrefList.append(a.get('href'))

hrefList
```

### 步骤3：原始的名词安装本身有类别，如下所示，你能将爬取的结果进行分类吗？

```python
secDict = defaultdict(list)

sections = soup.find(id="glossary-of-common-terms-and-api-elements").find_all('section')
for sec in sections:
    sec_id = sec.get('id')
    dts = sec.find_all('dt')
    for dt in dts:
        for st in dt.strings:
            secDict[sec_id].append(st)
            break

print(secDict)
    
```
